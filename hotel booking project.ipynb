```python
from scipy.stats import zscore
```

```python
column_name = ['adr', 'arrival_date_year']
threshold = 3  # Z-score threshold for outlier detection

# Loop through each column
for col in column_name:
    # Calculate z-scores for the column
    z_scores = zscore(newdf[col])

    # Identify outliers based on z-scores
    outliers = np.abs(z_scores) > threshold

    # Remove outliers from the DataFrame
    newdf = newdf[~outliers]
```

```python
plt.figure(figsize=(50, 12))
newdf.plot(kind='box')
plt.xticks(rotation=90)
plt.subplots_adjust(bottom=0.2)

plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (8, 4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = newdf, palette = 'Blues')
legend_handles, legend_labels = ax1.get_legend_handles_labels()
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotels')
plt.ylabel('number of reservation')
```

```python
resort_hotel = newdf[newdf['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = newdf[newdf['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
newdf.columns
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20, 8))
plt.title('Average Daily Rate in Resort and CIty Hotel', fontsize = 20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
```

```python
newdf['month'] = newdf['reservation_status_date'].dt.month
```

```python
plt.figure(figsize = (20, 8))
ax2 = sns.countplot(x = 'month', hue = 'is_canceled', data = newdf, palette = 'bright')
legend_handles, legend_labels = ax2.get_legend_handles_labels()
ax2.legend(bbox_to_anchor = (1,1))
plt.title('Reservaltion status per month', size = 20)
plt.xlabel('month')
plt.ylabel('No of reservation')
plt.legend('not_canceled', 'canceled')
plt.show()
```
